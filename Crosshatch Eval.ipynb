# Evaluating Crosshatch Router Performance

Evals provide a framework for evaluating large language models (LLMs) or systems built using LLMs. The framework can be found at https://github.com/openai/evals. It contains many evals created by community.

First, we need to pick an eval to run. The [registry](https://github.com/openai/evals/tree/main/evals/registry/evals) contains loads of evals to pick from. A few examples include:
- 2d_movement.dev.v0: Test the model's ability to keep track of position and orientation in a 2D environment.
- ambiguous-sentences.dev.v0: test pair of sentences that differ in only one or two words and that contain an ambiguity that is resolved in opposite ways in the two sentences and requires the use of world knowledge and reasoning for its resolution.
- emoji-riddle.s1.simple-v0: Test the model's ability to solve emoji riddles.
- finance.dev.v0: Test the model's ability to understand financial concepts and do math.
- game-theory.dev.v0: Practical reasoning about normal-form games such as Paper-Rock-Scissors
- heart-disease.v0: Test model's capability of predicting the presence of heart disease.
- irony.dev.v0: Tests the ability to identify one of three types of irony, situational, verbal, or dramatic.
- linear-equations.dev.v0
- MMLU (a common benchmark; multiple choice)
  - mmlu-business-ethics.val.ab-v1
  - mmlu-high-school-government-and-politics.val.ab-v1
  - mmlu-high-school-world-history.val.ab-v1
- qa.dev.v0: Tests the model's ability to correctly answer a question when the answer of the question is relocated in a great deal of context.
- invoices.dev.v0: Test the model's ability to parse real-world data, with an emphasis on numerical data

```python
import jsonlines
import pandas as pd
import os

from os import listdir
from os.path import isfile, join
```

### Get all logs...

```python
logs_filepath = "results/evallogs/"

files = [f for f in listdir(logs_filepath) if isfile(join(logs_filepath, f))]

for file in files:
    print(file)
```

### Analyze results

```python
# Let's get the filename of the latest run...

last_run_logs = ""

with open("results/runs.txt", 'r') as runs:
    last_line = runs.readlines()[-1]
    last_run_logs = last_line
    
print(f"Logs are here: {last_run_logs}")

with open(last_run_logs, "r") as f:
    events_df = pd.read_json(f, lines=True)


run_info = events_df.iloc[0]['spec']['run_config']
# print(events_df.iloc[0]['spec']['run_config'])
model = run_info['completion_fns'][0]
eval_spec = run_info['eval_spec']
# final_report = events_df.iloc[1]['final_report']

print(f"""Model: {model}
Eval-info: {eval_spec}
""")


matches_df = events_df[events_df.type == "match"].reset_index(drop=True)
matches_df = matches_df.join(pd.json_normalize(matches_df.data))


print(f"ABSOLUTE COUNTS:\n{matches_df.correct.value_counts()}")

print(f"\nPERCENTAGE:\n{matches_df.correct.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'}")

matches_df.correct.value_counts().plot.bar(title="Correctness of generated answers", xlabel="Correctness", ylabel="Count")
```

```python
# Inspect samples
for i, r in pd.json_normalize(events_df[events_df.type == "sampling"].data).iterrows():
    print(f"Prompt: {r.prompt[:100]}")
    print(f"Sampled: {r.sampled}")
    print("-" * 25)
```

## Let's compare against Crosshatch

```python
eval_selected = "qa.dev.v0"
num_samples = 100
model = "crosshatch_router" # Use crosshatch_router or use any OpenAI model: gpt-3.5-turbo, gpt-4-0125-preview, gpt-4 etc.

!oaieval {model} {eval_selected} --max_samples={num_samples}
```

```python
# Let's get the filename of the latest run...

last_run_logs = ""

with open("results/runs.txt", 'r') as runs:
    last_line = runs.readlines()[-1]
    last_run_logs = last_line
    
print(f"Logs are here: {last_run_logs}")

# last_run_logs = ""

with open(last_run_logs, "r") as f:
    events_df = pd.read_json(f, lines=True)


run_info = events_df.iloc[0]['spec']['run_config']
model = run_info['completion_fns'][0]
eval_spec = run_info['eval_spec']

print(f"""Model: {model}
Eval-info: {eval_spec}
""")


matches_df = events_df[events_df.type == "match"].reset_index(drop=True)
matches_df = matches_df.join(pd.json_normalize(matches_df.data))


print(f"ABSOLUTE COUNTS:\n{matches_df.correct.value_counts()}")

print(f"\nPERCENTAGE:\n{matches_df.correct.value_counts(normalize=True).mul(100).round(1).astype(str) + '%'}")

matches_df.correct.value_counts().plot.bar(title="Correctness of generated answers", xlabel="Correctness", ylabel="Count")
```

```python
# Inspect samples

for i, r in pd.json_normalize(events_df[events_df.type == "sampling"].data).iterrows():
    prompt = r["prompt.raw_prompt"]
    print(f"Prompt: {prompt}")
    print(f"Sampled: {r.sampled}")
    print("-" * 25)
```
